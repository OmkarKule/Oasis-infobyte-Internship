```python
# Option 1: Drop rows with missing values
df_cleaned = df.dropna()
```

```python
# Descriptive statistics for numeric columns
print(df_cleaned.describe())
```

```python
# Mode for categorical columns
print(df_cleaned.mode())
```

```python
# Group by category and calculate mean calories
category_summary = df_cleaned.groupby('Category')['Calories'].mean()
print(category_summary)

# Distribution of products in each category
category_counts = df_cleaned['Category'].value_counts()
print(category_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart for average calories by category
plt.figure(figsize=(10, 6))
sns.barplot(x=category_summary.index, y=category_summary.values)
plt.xlabel('Category')
plt.ylabel('Average Calories')
plt.title('Average Calories by Category')
plt.xticks(rotation=90)
plt.show()

# Heatmap of nutrient values (excluding rows with NaN)
nutrients = df_cleaned[['Total Fat', 'Saturated Fat', 'Trans Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']]
plt.figure(figsize=(12, 8))
sns.heatmap(nutrients.corr(), annot=True, cmap='coolwarm')
plt.title('Nutrient Correlations')
plt.show()
```
