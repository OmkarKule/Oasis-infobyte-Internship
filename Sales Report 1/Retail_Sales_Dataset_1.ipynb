```python
# Identify numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns
print("Numeric columns:", numeric_columns)

```

```python
# Calculate mean
mean_values = df[numeric_columns].mean()

# Calculate median
median_values = df[numeric_columns].median()

# Calculate mode
mode_values = df[numeric_columns].mode().iloc[0]

# Calculate standard deviation
std_dev_values = df[numeric_columns].std()

print("Mean values:\n", mean_values)
print("\nMedian values:\n", median_values)
print("\nMode values:\n", mode_values)
print("\nStandard Deviation values:\n", std_dev_values)
```

```python
import matplotlib.pyplot as plt

# Check if 'Date' column exists, if not, try another likely column name
if 'Date' not in df.columns:
    # Replace 'LikelyDateColumnName' with the potential name of your date column
    if 'LikelyDateColumnName' in df.columns:
        df.set_index('LikelyDateColumnName', inplace=True)
    else:
        print("Error: Neither 'Date' nor 'LikelyDateColumnName' column found in DataFrame.")
        # Handle the case where no date column is found, maybe load the data again?
else:
    df.set_index('Date', inplace=True)

# Proceed with resampling and plotting if a date column was found
if df.index.name in ['Date', 'LikelyDateColumnName']:  # Check if index was set successfully
    monthly_sales = df['Total Amount'].resample('M').sum()

    # Plot the monthly sales trend
    plt.figure(figsize=(12, 6))
    monthly_sales.plot(title='Monthly Sales Trend')
    plt.xlabel('Date')
    plt.ylabel('Total Sales')
    plt.grid(True)
    plt.show()
```

```python
# Analyze customer demographics
customer_age_distribution = df['Age'].value_counts()
gender_distribution = df['Gender'].value_counts()

# Analyze product categories
product_category_distribution = df['Product Category'].value_counts()

# Plot customer age distribution
plt.figure(figsize=(12, 6))
customer_age_distribution.plot(kind='bar', title='Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot gender distribution
plt.figure(figsize=(12, 6))
gender_distribution.plot(kind='bar', title='Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plot product category distribution
plt.figure(figsize=(12, 6))
product_category_distribution.plot(kind='bar', title='Product Category Distribution')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
