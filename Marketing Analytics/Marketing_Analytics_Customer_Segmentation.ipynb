```python
# Example: Fill missing values with the mean of the column
df.fillna(df.mean(), inplace=True)
```

```python
# Example: Check for negative values in columns that should only have positive values
for column in ['Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'MntTotal']:
    inconsistent_data = df[df[column] < 0]
    if not inconsistent_data.empty:
        print(f'Inconsistent data found in column {column}:')
        print(inconsistent_data)
        # Handle inconsistencies (e.g., replace negative values with 0 or drop rows)
        df[column] = df[column].apply(lambda x: max(x, 0))
```

```python
# Select relevant features for clustering
features = df[['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'MntTotal']]
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize the data
scaler = StandardScaler()
features_normalized = scaler.fit_transform(features)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Determine the optimal number of clusters using the Elbow method
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_normalized)
    sse.append(kmeans.inertia_)

# Plot the SSE for each k
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Apply K-means with the chosen number of clusters
optimal_k = 4  # This should be determined from the Elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(features_normalized)
```

```python
# Analyze the resulting clusters
cluster_analysis = df.groupby('Cluster').mean()
print(cluster_analysis)

# Visualize the clusters (optional)
import seaborn as sns

# Example: Visualize clusters based on Income and Total Amount Spent
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Income', y='MntTotal', hue='Cluster', palette='viridis')
plt.title('Customer Segments based on Income and Total Amount Spent')
plt.xlabel('Income')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of Income vs. Total Amount Spent, colored by cluster
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Income', y='MntTotal', hue='Cluster', palette='viridis')
plt.title('Customer Segments based on Income and Total Amount Spent')
plt.xlabel('Income')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
# Calculate the mean values for each cluster
cluster_means = df.groupby('Cluster').mean().reset_index()

# Bar chart of mean Income by cluster
plt.figure(figsize=(10, 6))
sns.barplot(data=cluster_means, x='Cluster', y='Income', palette='viridis')
plt.title('Average Income by Customer Segment')
plt.xlabel('Cluster')
plt.ylabel('Average Income')
plt.show()

# Bar chart of mean Total Amount Spent by cluster
plt.figure(figsize=(10, 6))
sns.barplot(data=cluster_means, x='Cluster', y='MntTotal', palette='viridis')
plt.title('Average Total Amount Spent by Customer Segment')
plt.xlabel('Cluster')
plt.ylabel('Average Total Amount Spent')
plt.show()
```

```python
# Box plot of Income by cluster
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Cluster', y='Income', palette='viridis')
plt.title('Income Distribution by Customer Segment')
plt.xlabel('Cluster')
plt.ylabel('Income')
plt.show()

# Box plot of Total Amount Spent by cluster
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Cluster', y='MntTotal', palette='viridis')
plt.title('Total Amount Spent Distribution by Customer Segment')
plt.xlabel('Cluster')
plt.ylabel('Total Amount Spent')
plt.show()
```
